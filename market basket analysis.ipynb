```python
market = pd.read_csv('../input/the-bread-basket/bread basket.csv')
market.head(10)
```

```python
Basket =  market.groupby(['Transaction','Item'])['Transaction'].sum().unstack().reset_index().fillna(0).set_index('Transaction')
```

```python
Basket.head(10)
```

```python
Basket.shape
```

```python
# Convert nilai menjadi 0 dan 1

def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
Basket.columns
```

```python
Basket_sets = Basket.applymap(encode_units)
```

```python
Basket_sets.head()
```

```python
Basket_sets.shape
```

```python
frequent_itemsets = apriori(Basket_sets, min_support=0.05, use_colnames=True)
```

```python
frequent_itemsets.shape
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=0.003)
rules.head()
```

```python
rules.shape
```

```python
rules[ (rules['lift'] >= 0.05) &
       (rules['confidence'] >= 0.5) ]
```

```python
#melihat item cakes dan coffie yang terjual
cakes = Basket['Cake'].sum()
coffees = Basket['Coffee'].sum()
print(f'Total cakes sold: {cakes} and total coffee sold: {coffees}')
```
